**ONE HOT ENCODING ON MULTIPLE CATEGORIES**

```python
#providing the path of the file and storing the data in dataset and pandas library is used to study the data
df= pd.read_csv('/content/drive/MyDrive/Colab Notebooks/mercedesbenz.csv',usecols=['X0','X1','X2','X3','X4','X5','X6'])
```

```python
df.head()
```

```python
for i in df.columns:
  print(df[i].value_counts())
```

```python
for i in df.columns:
  print(df[i].unique())
```

```python
for i in df.columns:
  print(len(df[i].unique()))
```

```python
#kdd cup orange challenge with ensemble technique
#take first top 10 categories andapply one hot encoding on top 10 categories
```

```python
df.X1.value_counts().sort_values(ascending=False).head(10)
```

```python
df.X1.value_counts().sort_values(ascending=False).head(10).index
```

```python
lst_10=df.X1.value_counts().sort_values(ascending=False).head(10).index
lst_10=list(lst_10)
```

```python
lst_10
```

```python
import numpy as np
for categories in lst_10:
  df[categories]=np.where(df['X1']==categories,1,0)
```
