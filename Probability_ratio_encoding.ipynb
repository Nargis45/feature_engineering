# **Probability ratio encoding**

Step1: probability of survived based on cabin

Step2: probability of not survived

Step3: prob(survived)/prob(not survived)

Step4: dictionary to map cabin with probability

Step5: replace with the categorical fature

```python
prob_df=df.groupby(['Cabin'])['Survived'].mean()
```

```python
prob_df=pd.DataFrame(prob_df)
prob_df
```

```python
#prob. person has died
prob_df['died']=1-prob_df['Survived']
```

```python
prob_df
```

```python
#prob. ratio encoding
prob_df['prob_ratio']=prob_df['Survived']/prob_df['died']
```

```python
prob_df
```

```python
prob_encoded=prob_df['prob_ratio'].to_dict()
```

```python
df['Cabin_encoded']=df['Cabin'].map(prob_encoded)
```
